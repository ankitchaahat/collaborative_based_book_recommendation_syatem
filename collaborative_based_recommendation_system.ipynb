# Popularity Based

```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num-ratings'},inplace=True)
num_rating_df
```

```python
num_rating_df.info()
```

```python
# Convert 'Book-Rating' column to numeric, handling non-numeric values
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')

avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg-ratings'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num-ratings']>=250].sort_values('avg-ratings',ascending=False)
```

```python
popular_df
```

```python
popular_df.merge(books,on='Book-Title')
```

```python
popular_df = popular_df.drop_duplicates(subset='Book-Title',keep=False) # Use the correct function name to drop duplicates based on 'Book-Title' column.
```

```python
popular_df
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num-ratings','avg-ratings']]
```

# Collaborativr Filtering Based

```python
ratings_with_name
```

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']  # number of people rated
x
```

```python
# number of people rated more than 200 ratings

x=x[x>200]
x
```

```python
# Get the index of users who have rated more than 200 books
popular_users = x[x > 200].index
popular_users
```

```python
filtered_ratings=ratings_with_name[ratings_with_name['User-ID'].isin(popular_users)]
```

```python
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
y
```

```python
famous_books=y[y].index
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
  # fetch index
  index=np.where(pt.index==book_name)[0][0]
  similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]

  for i in similar_items:
    print(pt.index[i[0]])
```

```python
def recommend(book_name):
  # fetch index
  index=np.where(pt.index==book_name)[0][0]
  similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]

  data=[]
  for i in similar_items:
    item=[]
    # Check if 'Book-Title' is the correct column name, adjust if needed
    temp_df=books[books['Book-Title']==pt.index[i[0]]]
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

    data.append(item)
  return data
```
